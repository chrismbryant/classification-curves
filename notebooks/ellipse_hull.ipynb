```python
from matplotlib.patches import Ellipse
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt

from clscurves.covariance import EllipsePairHull

# Define the ellipses
e1 = Ellipse(xy=(0, 1), width=4, height=8, angle=0)
e2 = Ellipse(xy=(4, 4), width=4, height=8, angle=45)

eph = EllipsePairHull(e1, e2)
points, hull = eph.get_points_and_hull()
hull = points[hull]
segments = eph.get_hull_segments()

# Plot patches
fig, ax = plt.subplots(dpi=150)
ax.add_patch(e1)
ax.add_patch(e2)
ax.scatter(hull[:,0], hull[:,1], color="black")

# Plot segments
for segment in segments:
    ax.plot(
        segment[:, 0],
        segment[:, 1],
        color="red",
    )

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
```
